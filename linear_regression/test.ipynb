```python
from collections import defaultdict
class Solution:
    def finalPrices(self, prices: List[int]) -> List[int]:
        stack = []
        res = prices.copy()
        for i, a in enumerate(prices):
            while stack and prices[stack[-1]] >= a:
                res[stack[-1]] = prices[stack[-1]] - a
                stack.pop()
            stack.append(i)
        return res
    



Solution().finalPrices([1,2,3,4,5])
```
